<h1> Dataset </h1>
<body> A dataset is a collection of data. It can be in the form of a table, images, sql, no-sql, text corpus, etc. </body>
<body> After a data is collected from the source, it is called as <b> raw data </b>. Raw data consists of many issues like missing values, noise, outliers, etc. To clean this data, <b> IDA </b> or <b> Initial data analysis </b> is performed. After the data is cleaned, a data scientist will try to <b> TALK TO THE DATA </b>. We try to understand and investigate the hidden patterns, messages and clues in the data. This step is called <b> Exploratory Data Analysis or EDA </b>.

<h1> Initial Data Analysis </h1>
<body> Initial data analysis refers to the series of steps used to clean the raw data. It consists of steps like missing value imputations, data revision, hypothesis testing, power analysis and some other statistical steps.</body>

<h1> Exploratory Data Analysis (EDA) </h1>
<body> In statistics, <b> Exploratory Data Analysis </b> is an approach of analyzing datasets to summarize their main characteristics, often using statistical graphics and other data visualization methds. Primarily, EDA is used for seeing what the data can tell us beyond the formal modelling or hypothesis testing* task. </body>

<h1> The IRIS dataset </h1>

    <p> It is a dataset that is used to identify three species of flower belonging to the Iris genus.</p>
        <ul>
            <li> Iris setosa </li>
            <li> Iris versicolor </li>
            <li> Iris virginica </li>
    </ul>

```python
# Here we will be importing pandas.
# Pandas is a python library which is used to read and perform various operations on datasets.
# It is painfully slow sometimes but it is the most o=commonly used.
import pandas as pd
dataset = pd.read_csv('Iris.csv')
```

```python
dataset.head(10)
```

<body>
    <p> As we can see, the dataset consists of 6 columns : 
        <ul>
            <li> Id - Unique value for each observation. </li>
            <li> SepalLengthCm - The length of the flower sepal in centimetres. </li>
            <li> SepalWidthCm - the width of the flower sepal in centimetres </li>
            <li> PetalLengthCm - The length of the flower petal in centimetres </li>
            <li> PetalWidthCm - The width of the petal in centimetres </li>
            <li> Species - The species to which the observed flower belongs to. </li>

<h3>So, what are these columns? </h3>
<body>
    <p> The column "Id" is nothing but a unique identifier of the observations. It is like the serial number for each observations. As one can easily think, this column will not give us any meaningful insight into the data.</p>
    <p> The columns "SepalLengthCm" , "SepalWidthCm" , "PetalLengthCm" , "PetalWidthCm" are called as feature columns or "Features". These columns contain the features of the flower based on which a botanist differentiates between them. In this scenario, we will try to visualize these features and try to deduce meaningful insights from them. </p>
    <p> The column "Species" contains the information about the species that the observed flower belongs to. This is column that our machine learning algorithm will try to predict. </p>

```python
# Lets first see the amount of data that we have to work with. 
# Here we will be answering the question of how many data points and how many features ?
print(dataset.shape)
```

```python
# Since  don't need the Id column, so we will be dropping it
dataset = dataset.drop("Id" , axis = 1)
```

```python
#Let's see the dataset after dropping the Id column
print(dataset.shape)
dataset.head(10)
```

```python
# Now that we have seen the data, we can ask the question
# how many datapoints form each class are present in the dataset ?
dataset['Species'].value_counts()
```

<p> The dataset has 50 observations from each class. These kinds of datasets where the number of observations form each class are the same is called as a "BALANCED DATASET". The dataset in which the number of observations form each class differ widley, is called as an "IMBALANCED DATASET". </p>

<p> Imbalanced datasets are very normal in the real world but they are problematic to work on because computers are stupid. </p>

<h2> Let's visualize the data !!!! </h2>
<body>
    <p> Data visualization is the secret weapon of data scientists. Data holds the key to everything. If we aks the correct questions, we will get the correct answers. To be able to ask the correct questions, we should first understand what the data is saying. </p>

```python
# Lets import the libraries required to visualize the data
import warnings
#current version of the plotting library seaborn generates a lot of warning which we are going to ignore
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style = "white" , color_codes = True)
```

<h3> 2D Scatter Plot </h3>

```python
dataset.plot(kind = "scatter" , x = "SepalLengthCm" , y = "SepalWidthCm")
plt.show()

# We will ignore the angry red texts because sometimes computers are moody and stupid.
```

<h2> Observations </h2>
<body>
    <ul>
        <li>The sepal length of the flowers vary from 4 to 8 cm</li>
        <li> The sepal width of the flowers vary from 2 to 4.5 cm </li>
        <li> Except for these basic observations about the general size of the entire genus, we cannot deduce anything else from this scatter plot </li>
    </ul>
    <p> So let's see what happens when we color code the points according to the variosu species or classes </p>
    </body>

```python
# We will be using seaborn for plotting because it is fircking awesome !!!!!!

sns.FacetGrid(dataset , hue = "Species" , size = 5)\
.map(plt.scatter, "SepalLengthCm" , "SepalWidthCm")\
.add_legend()
```

And suddenly the data is telling us a lot of things !!!!!!!
<h3> Observations </h3>
<body>
    <ul>
        <li> The blue points representing the iris setosa species can be easily separated form the others just by drawing a line, hence it is linearly separable. </li>
        <li> The orange and green data points representing the "iris versicolor" and "iris virginica" species cannot be separated from each other easily as they have a considerable overlap. </li>
    </ul>
    <p> So now some inquisitive souls among youall might ask, if we can plot the datapoints based on 2 features, can we use other features to see whether we get a better idea about the data ? </p>
    <p> And the answer is YES !! We can take any 2 features and create a scatter plot. the total number of scatter plots that can be created form these 4 features is 4C2 = 6 </p>

<h3> Pair Plot </h3>

```python
sns.pairplot(dataset , hue = "Species" , size = 3)
```

Note:
    "For now, we will be ignoring the scary looking diagonal plots. We will be focussing on out cute little scatter plots "

<h3> Observations : </h3>
<body>
    <ul>
        <li> Petal Width and Petal Length are probably the most useful features to identify various types of flowers. </li>
        <li>While setosa can be easily identified (linearly separable), Virginica and Versicolor have some degree of overlap (almost linearly separable).</li>
        <li>  We can easily find the class boundaries (lines that separate the points from each other) and create a if-else model to classify the species of flowers.</li>
    </ul>
    <p> NOW THE DATA IS SURELY SPILLING SOME OF ITS SECRETS !! </p>

<h2> Univariate analysis or Descriptive Analysis using PDF </h2>
<body>
    <p>Descriptive analysis, or univariate analysis, provides an understanding of the characteristics of each attribute of the dataset.</p>
    <p> Here we will see how we can use the simple comcept of a histogram of a probability density function to understand the data. </p>
    <p> Univariate Analysis means analysis of the data one variable at a time. </p>

```python
#here we will be plotting a histogg for petal length

sns.FacetGrid(dataset , hue = "Species" , size = 5)\
.map(sns.distplot , "PetalLengthCm")\
.add_legend()
```

<h3> Observations with petal length </h3>
<body>
    <ul>
        <li> Setosa is well separated form versicolor and virginica on the basis of petal length. There are no setosa flowers beyon petal length of 2 cm </li>
        <li> Between versicolor and virginica, there is a small overlap. </li>

```python
#here we will be plotting a histogram for petal width

sns.FacetGrid(dataset , hue = "Species" , size = 5)\
.map(sns.distplot , "PetalWidthCm")\
.add_legend()
```

<h3> Observations using petal width </h3>
<body>
    <ul>
        <li> It can be noticed that setosa is a bit closer to versicolor that it was in the case of petal length.</li>
        <li> Virginica and versicolor have almost the same overlap. But the distributions are more wide. So, petal length is a better feature.</li>

```python
#here we will be plotting a histogram for sepal length

sns.FacetGrid(dataset , hue = "Species" , size = 5)\
.map(sns.distplot , "SepalLengthCm")\
.add_legend()
```

<h3> Observations with sepal length </h3>
<body>
    <ul>
        <li> This is scary </li>
        <li> There are huge over laps between all the distributions. </li>
        <li> Its the kind of distribution your data tells you to stay away from </li>

```python
#here we will be plotting a histogram for sepal width

sns.FacetGrid(dataset , hue = "Species" , size = 5)\
.map(sns.distplot , "SepalWidthCm")\
.add_legend()
```

<h3> Obervations with sepal width </h3>
<body>
    <ul>
        <li> This is even worse that sepal length.</li>
        <li> Virginical and versicolor are almost indistinguishable from each other. </li>
        <li> There is also no significant difference betwween setosa and the others. </li>
    </ul>
    <p> So, we can conlclude from the univariate analysis of the data that petal length > petal width >> sepal length >> sepal width
